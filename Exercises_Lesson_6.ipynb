### 1. Read the different files with Cyprus towns temperatures provided in the TData folder and build a dataframe combining all the information. The columns should be the  year and months and you can distinguish between data for the different towns adding an extra column with the town name. Hint: the function concat can be very useful in this case.

```python
months = ["Jan", "Feb", "Mar", "Apr","May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
years  = np.arange(1961, 2096+1, 1)
lis_y  = list(np.array([[years[i]]*12 for i in range(len(years))]).ravel())
```

```python
tdata_Agrinio = pd.read_csv("./TData/T_Agrinio_EM.csv", index_col=0, names = months, skiprows=1)
tdata_Ajjacio = pd.read_csv("./TData/T_Ajjacio_EM.csv", index_col=0, names = months, skiprows=1)
tdata_Alexandroupolis = pd.read_csv("./TData/T_Alexandroupolis_EM.csv", index_col=0, names = months, skiprows=1)
tdata_Alicante = pd.read_csv("./TData/T_Alicante_EM.csv", index_col=0, names = months, skiprows=1)
tdata_Antalya = pd.read_csv("./TData/T_Antalya_EM.csv", index_col=0, names = months, skiprows=1)
tdata_Araxos = pd.read_csv("./TData/T_Araxos_EM.csv", index_col=0, names = months , skiprows=1)
```

```python
dframe_Agrinio = pd.DataFrame([tdata_Agrinio.values.ravel()], columns=[lis_y,months*136], index=[""])
dframe_Ajjacio = pd.DataFrame([tdata_Ajjacio.values.ravel()], columns=[lis_y,months*136], index=[""])
dframe_Alexandroupolis = pd.DataFrame([tdata_Alexandroupolis.values.ravel()], columns=[lis_y,months*136], index=[""])
dframe_Alicante = pd.DataFrame([tdata_Alicante.values.ravel()], columns=[lis_y,months*136], index=[""])
dframe_Antalya = pd.DataFrame([tdata_Antalya.values.ravel()], columns=[lis_y,months*136], index=[""])
dframe_Araxos = pd.DataFrame([tdata_Araxos.values.ravel()], columns=[lis_y,months*136], index=[""])
```

```python
concat = pd.concat([dframe_Agrinio, dframe_Ajjacio, dframe_Alexandroupolis,dframe_Alicante, dframe_Antalya,dframe_Araxos],
          keys= ["Agrinio","Ajjacio", "Alexandroupolis", "Alicante", "Antalya", "Araxos" ])
concat.index.names = ["Towns", ""]
concat
```

### 2. A different way to combine the Cyprus towns temperature data provided in the TData folder is to build a dataframe whose index is a hierarchical one, with the year and month and there are as many columns as towns, labeled with the town names. Hint: The concat and unstack functions can be helpful in this case. 

```python
conc = pd.concat([tdata_Agrinio.stack(), tdata_Ajjacio.stack(), tdata_Alexandroupolis.stack(),tdata_Alicante.stack(), tdata_Antalya.stack(), tdata_Agrinio.stack()],
                 axis=1, keys = ["Agrinio", "Ajjacio","Alexandroupolis","Alicante","Antalya","Agrinio"])
dframe2= pd.DataFrame(conc)
dframe2.index.names = ["Years", "Month"]
dframe2
```
