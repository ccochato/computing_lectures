### 1. Read the different files with Cyprus towns temperatures provided in the TData folder and build a dataframe combining all the information. The columns should be the  year and months and you can distinguish between data for the different towns adding an extra column with the town name. Hint: the function concat can be very useful in this case.

```python
months = ["Jan", "Feb", "Mar", "Apr","May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
years  = np.arange(1961, 2096+1, 1)
lis_y  = list(np.array([[years[i]]*12 for i in range(len(years))]).ravel())
```

```python
tdata_Agrinio = pd.read_csv("./TData/T_Agrinio_EM.csv", index_col=0, names = months, skiprows=1)
tdata_Ajjacio = pd.read_csv("./TData/T_Ajjacio_EM.csv", index_col=0, names = months, skiprows=1)
tdata_Alexandroupolis = pd.read_csv("./TData/T_Alexandroupolis_EM.csv", index_col=0, names = months, skiprows=1)
tdata_Alicante = pd.read_csv("./TData/T_Alicante_EM.csv", index_col=0, names = months, skiprows=1)
tdata_Antalya = pd.read_csv("./TData/T_Antalya_EM.csv", index_col=0, names = months, skiprows=1)
tdata_Araxos = pd.read_csv("./TData/T_Araxos_EM.csv", index_col=0, names = months , skiprows=1)
```

```python
dframe_Agrinio = pd.DataFrame([tdata_Agrinio.values.ravel()], columns=[lis_y,months*136])
dframe_Ajjacio = pd.DataFrame([tdata_Ajjacio.values.ravel()], columns=[lis_y,months*136])
dframe_Alexandroupolis = pd.DataFrame([tdata_Alexandroupolis.values.ravel()], columns=[lis_y,months*136])
dframe_Alicante = pd.DataFrame([tdata_Alicante.values.ravel()], columns=[lis_y,months*136])
dframe_Antalya = pd.DataFrame([tdata_Antalya.values.ravel()], columns=[lis_y,months*136])
dframe_Araxos = pd.DataFrame([tdata_Araxos.values.ravel()], columns=[lis_y,months*136])
```

```python
concat = pd.concat([dframe_Agrinio, dframe_Ajjacio, dframe_Alexandroupolis,dframe_Alicante,
                    dframe_Antalya,dframe_Araxos])
concat.index = ["Agrinio","Ajjacio", "Alexandroupolis", "Alicante", "Antalya", "Araxos" ]
concat.index.names = ["Towns"]
concat
```

### 2. A different way to combine the Cyprus towns temperature data provided in the TData folder is to build a dataframe whose index is a hierarchical one, with the year and month and there are as many columns as towns, labeled with the town names. Hint: The concat and unstack functions can be helpful in this case. 

```python
conc = pd.concat([tdata_Agrinio.stack(), tdata_Ajjacio.stack(), tdata_Alexandroupolis.stack(),tdata_Alicante.stack(), tdata_Antalya.stack(), tdata_Agrinio.stack()],
                 axis=1, keys = ["Agrinio", "Ajjacio","Alexandroupolis","Alicante","Antalya","Agrinio"])
dframe2= pd.DataFrame(conc)
dframe2.index.names = ["Years", "Month"]
dframe2
```

### 3. Compute the correlation matrix between the temperatures in the provided Cyprus cities from the previous exercise dataframe.

```python
print(dframe2.corr().values)
```

### 4. We provide as  an example data set the file meteodat.csv with an excerpt of data with a 10 minute frequency from an automated meteorological station with a span of two months (Jan and Feb 2014). Data are comma separated and you can read them using pd.read_csv. The first column is the date and the second the time.  Transform this data into a dataframe with an index of datetime.

```python
tdata_meteodat = pd.read_csv("./meteodat.csv")
headers = list(tdata_meteodat.columns.values)[2:]
date = tdata_meteodat.values[0:,0].ravel()
time = tdata_meteodat.values[0:,1].ravel()
datetime = [date[i]+ str(" ")+time[i] for i in range(len(time))]
array = tdata_meteodat.values[0:,2:]
```

```python
dframe_meteodat = pd.DataFrame(array, columns = headers,  index= datetime)
dframe_meteodat.index = pd.to_datetime(dframe_meteodat.index)
dframe_meteodat.index.names = ["Date-Time"]
dframe_meteodat
```

###  Compute dataframes with downsampling to hourly and daily average values for temperature (Tout), pressure (Pressure), relative humidity (H out), wind speed (Wind Speed), and dew point temperature (Dew point); 

```python
new_meteodat = dframe_meteodat[['Tout','Pressure','H out', 'Wind Speed','Dew point']]
new_meteodat_hourly = new_meteodat.resample("1H").sum()
new_meteodat_hourly
```

```python
new_meteodat_daily = new_meteodat.resample("24H").sum()
new_meteodat_daily
```
