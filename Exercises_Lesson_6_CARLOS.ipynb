```python
# Creating useful variables
months = ["Jan", "Feb", "Mar", "Apr","May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
towns = ["Agrinio",'Ajjacio', 'Alexandroupolis', 'Alicante', 'Antalya', 'Araxos'] 
```

## 1. Read the different files with Cyprus towns temperatures provided in the TData folder and build a dataframe combining all the information. The columns should be the  year and months and you can distinguish between data for the different towns adding an extra column with the town name. Hint: the function concat can be very useful in this case.

```python
data = []
for town in towns:
    # Read the different files with Cyprus towns temperatures
    city = pd.read_csv("./TData/T_"+ town+ "_EM.csv", index_col = 0, names = months, skiprows = 1)
    # Add an extra column with the town name
    city['Town']= town
    data.append(city)
conc = pd.concat(data)
dframe = pd.DataFrame(conc)
dframe
```

## 2. A different way to combine the Cyprus towns temperature data provided in the TData folder is to build a dataframe whose index is a hierarchical one, with the year and month and there are as many columns as towns, labeled with the town names. Hint: The concat and unstack functions can be helpful in this case. 

```python
data = []
for town in towns:
    # Read the different files with Cyprus towns temperatures
    city = pd.read_csv("./TData/T_"+ town+ "_EM.csv", index_col = 0, names = months, skiprows = 1)
    # Calculate the transpose and unstack the city data
    data.append(city.T.unstack())
conc = pd.concat(data, axis = 1, keys = towns)
dframe2 = pd.DataFrame(conc)
dframe2.index.names = ["Years", "Month"]
dframe2
```

## 3. Compute the correlation matrix between the temperatures in the provided Cyprus cities from the previous exercise dataframe.

```python
print(dframe2.corr().values)
```

## 4. We provide as  an example data set the file meteodat.csv with an excerpt of data with a 10 minute frequency from an automated meteorological station with a span of two months (Jan and Feb 2014). Data are comma separated and you can read them using pd.read_csv. The first column is the date and the second the time.  Transform this data into a dataframe with an index of datetime.

```python
# Reading the data and creating useful varibles
tdata_meteodat = pd.read_csv("./meteodat.csv")
headers = list(tdata_meteodat.columns.values)[2:]
date = tdata_meteodat.values[0:, 0].ravel()
time = tdata_meteodat.values[0:, 1].ravel()
datetime = [date[i]+ str("-")+time[i] for i in range(len(time))]
array = tdata_meteodat.values[0:, 2:]
```

```python
# Transform this data into a dataframe with an index of datetime
dframe_meteodat = pd.DataFrame(array, columns = headers,  index = datetime)
dframe_meteodat.index = pd.to_datetime(dframe_meteodat.index, dayfirst = True)
dframe_meteodat.index.names = ["Date-Time"]
dframe_meteodat
```

###  Compute dataframes with downsampling to hourly and daily average values for temperature (Tout), pressure (Pressure), relative humidity (H out), wind speed (Wind Speed), and dew point temperature (Dew point); 

### Dataframe with downsampling to hourly average values

```python
new_meteodat = dframe_meteodat[['Tout','Pressure','H out', 'Wind Speed','Dew point']].astype(float)
new_meteodat_hourly = new_meteodat.resample("1H").mean()
new_meteodat_hourly
```

#### hourly  correlation

```python
print(new_meteodat_hourly.corr())
```

### Dataframe with downsampling to daily average values.

```python
new_meteodat_daily = new_meteodat.resample("1D").mean()
new_meteodat_daily
```

#### Daily Correlation

```python
print(new_meteodat_daily.corr())
```
